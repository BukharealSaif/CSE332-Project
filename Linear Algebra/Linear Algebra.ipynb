```python
import numpy as np
```

```python
a = np.array([[10, 15]])
b = np.array([[20, 9]])
print(a)
print(b)
print()

(a + b).T
```

```python
#Subtracting 2 vector

(a - b).T
```

```python
#Scaler

a = 2
u = np.array([[30, 15]])

(a * u).T
```

```python
#Dot product of 2 vector
#For 2-D vectors, its the equivalent to matrix multiplication

u = np.array([10, 20])
v = np.array([25, 6])

print(10 * 25 + 20 * 6)

np.dot(u,v)
```

```python
#Matrix Addition

a = np.array([[12, 10], [4, 6], [40, 31]])
b = np.array([[2, 8], [44, 63], [50, 10]])

print(a, end=" =a\n\n")
print(b, end=" =b\n\n")

a + b
```

```python
#Matrix Subtraction

a-b
```

```python
#Matrix scaler multiplication

a = 4
b = np.array([[2,5], [3, 7], [8, 6],])

a * b
```

```python
#Matrix & vextor multiplication

a = [[5,10], [15,20]]
b = [10, 12]

np.dot(a,b)
```

```python
#Matrix to matrix multiplication

a = [[5,10], [15,20], [25, 30]]
b = [[3,6], [9,5]]

np.dot(a, b)
```

```python
#Transpose a matrix

a = [[5,10], [15,20], [25, 30]]
print(a, end=" =a\n\n")
```

```python
#Create an identity matrix

I1 = np.eye(2)
I2 = np.eye(3)

print(I1, end=" = I1\n\n")
print(I2, end=" = I2\n\n")
```

```python
#Inverse a matrix

a = np.array([
    [2,1,0],
    [0,1,0],
    [1,2,1],
])
print(a, end=" =a\n\n")
print(np.linalg.inv(a), end=" = |a|")
```
