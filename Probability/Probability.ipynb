```python
from itertools import product as prod

A = set([x for x in range(1,7)])
B = set([x for x in range(1,7)])
p=list(prod(A,B))

print("A is set of all possible outcomes of a dice:",A)
print("B is set of all possible outcomes of a dice:",B)
print("\nProduct of A and B is set of all possible combinations of A and B thrown together:\n",p)
```

```python
#Throwing 2 dice

n_dice = 2
dice_faces = {1,2,3,4,5,6}

#Construct the event space i.e. set of ALL POSSIBLE events
event_space= set(prod(dice_faces,repeat=n_dice))
event_space
```

```python
for outcomes in event_space:
  print(outcomes, end=', ')
print()
print(len(event_space))  
```

```python
#Sum is a multiple of 3

favorable_outcomes =[]
for outcomes in event_space:
  x,y = outcomes
  if (x+y)%3==0:
    favorable_outcomes.append(outcomes)
favorable_outcomes = set(favorable_outcomes)  

for f_outcome in favorable_outcomes:
  print(f_outcome, end=', ')

print()
print(len(favorable_outcomes))

prob = len(favorable_outcomes)/len(event_space)
print("The probability of getting a sum which is a multiple of 3 is: ", prob)
```

```python
#Sum is a multiple of 5 but not a multiple of 3

favorable_outcomes =[]
for outcomes in event_space:
  d1,d2,d3,d4,d5 = outcomes
  if(d1+d2+d3+d4+d5)%5==0 and (d1+d2+d3+d4+d5)%3!=0:
    favorable_outcomes.append(outcomes)

favorable_outcomes = set(favorable_outcomes)  
len(favorable_outcomes) 
```

```python
prob = len(favorable_outcomes)/len(event_space)
print("The probability of getting a sum which is a multiple of 5 but not a multiple of 3 is: ", prob)
```
